```python
x=data.drop("Price",axis=1)
```

```python
y=data.Price
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,root_mean_squared_error
```

```python
mean_absolute_error(y_test,model.predict(x_test))
```

```python
mean_squared_error(y_test,model.predict(x_test))
```

```python
r2_score(y_test,model.predict(x_test))
```

```python
root_mean_squared_error(y_test,model.predict(x_test))
```

```python
from sklearn.model_selection import cross_val_score,KFold
```

```python
k=KFold(n_splits=3)
```

```python
cross_val_score(LinearRegression(),x_train,y_train,cv=k)
```

```python
cross_val_score(KNeighborsRegressor(),x_train,y_train,cv=k)
```

```python
cross_val_score(RandomForestRegressor(),x_train,y_train,cv=k)
```

```python
cross_val_score(DecisionTreeRegressor(),x_train,y_train,cv=k)
```

```python
cross_val_score(SVR(),x_train,y_train,cv=k)
```

```python
modelRF=RandomForestRegressor(n_estimators=200,random_state=20)
```

```python
modelRF.fit(x_train,y_train)
```

```python
modelRF.score(x_train,y_train)
```

```python
modelRF.score(x_test,y_test)
```

```python
mean_absolute_error(y_test,modelRF.predict(x_test))
```

```python
mean_squared_error(y_test,modelRF.predict(x_test))
```

```python
r2_score(y_test,modelRF.predict(x_test))
```

```python
root_mean_squared_error(y_test,modelRF.predict(x_test))
```
